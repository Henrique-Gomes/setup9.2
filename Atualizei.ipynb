```python
#funções
#calculo MME9
def MME9(df,aux,tipo):
    for i in range(df.shape[0]):
        aux[i].drop(columns=['MME9','SLOPE'],errors='ignore',inplace = True)
        aux[i].insert(aux[i].shape[1],'MME9',aux[i].Close.ewm(span=9).mean().dropna()) 
        aux[i].insert(aux[i].shape[1],'SLOPE',0.0)
        for j in range (1,aux[i].shape[0]):
            tend = aux[i].MME9[j] - aux[i].MME9[j-1]
            aux[i].iat[j,aux[i].shape[1]-1]=tend
        df.iat[i,tipo] = aux[i];
    
    return(df,aux)

def tendencia(df,aux,tipo):
    #toposefundos
    fundos = []
    topos = []
    for i in range(df.shape[0]):
        aux[i]['TOPO'] = ''
        aux[i]['FUNDO'] = ''
        inferno1 = []
        inferno2 = []
        j=0;
        for k in aux[i].index:
            if j == aux[i].shape[0]-1:
                break
            else:
                aux[i].loc[k,'TOPO'] = np.where((aux[i].High[j-1] < aux[i].High[j] > aux[i].High[j+1]),1,0)
                aux[i].loc[k,'FUNDO'] = np.where((aux[i].Low[j-1] > aux[i].Low[j] < aux[i].Low[j+1]),1,0)
                j += 1;
            
        aux[i].at[aux[i].index[-1],'TOPO'] = np.array(0)
        aux[i].at[aux[i].index[0],'TOPO'] = np.array(0)
        aux[i].at[aux[i].index[-1],'FUNDO'] = np.array(0)
        aux[i].at[aux[i].index[0],'FUNDO'] = np.array(0)
        
        for j in aux[i].index:
            if int(aux[i].at[j,'TOPO']) == 1:
                inferno1.append(aux[i].High[j]);
            if int(aux[i].at[j,'FUNDO']) == 1:
                inferno2.append(aux[i].Low[j]);
        topos.append(inferno1)
        fundos.append(inferno2)
        df.iat[i,tipo] = aux[i];        
            #regras de tendencia
    if tipo == 1:
        for i in range(0,len(topos)-1):
            if ((len(fundos[i])>2) & (len(topos[i])>2)):
                if (topos[i][-1] > topos[i][-2]) & (fundos[i][-1] > fundos[i][-2]):
                    df['TENDENCIA'][i] = 'ALTA'
                if (topos[i][-1] < topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]):
                    df['TENDENCIA'][i] = 'BAIXA'
                if (((topos[i][-1] < topos[i][-2]) & (fundos[i][-1] > fundos[i][-2])) | ((topos[i][-1] > topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]))):
                    df['TENDENCIA'][i] = 'INDEFINIDA'

            else:
                df['TENDENCIA'][i] = 'INDEFINIDA'
    if tipo == 2:
        for i in range(0,len(topos)-1):
            if ((len(fundos[i])>2) & (len(topos[i])>2)):
                if (topos[i][-1] > topos[i][-2]) & (fundos[i][-1] > fundos[i][-2]):
                    df['TENDENCIAWK'][i] = 'ALTA'
                if (topos[i][-1] < topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]):
                    df['TENDENCIAWK'][i] = 'BAIXA'
                if (((topos[i][-1] < topos[i][-2]) & (fundos[i][-1] > fundos[i][-2])) | ((topos[i][-1] > topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]))):
                    df['TENDENCIAWK'][i] = 'INDEFINIDA'

            else:
                df['TENDENCIAWK'][i] = 'INDEFINIDA'
                
    if tipo == 3:
        for i in range(0,len(topos)-1):
            if ((len(fundos[i])>2) & (len(topos[i])>2)):
                if (topos[i][-1] > topos[i][-2]) & (fundos[i][-1] > fundos[i][-2]):
                    df['TENDENCIAMO'][i] = 'ALTA'
                if (topos[i][-1] < topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]):
                    df['TENDENCIAMO'][i] = 'BAIXA'
                if (((topos[i][-1] < topos[i][-2]) & (fundos[i][-1] > fundos[i][-2])) | ((topos[i][-1] > topos[i][-2]) & (fundos[i][-1] < fundos[i][-2]))):
                    df['TENDENCIAMO'][i] = 'INDEFINIDA'

            else:
                df['TENDENCIAMO'][i] = 'INDEFINIDA'
    return(df,aux)
  
    
def atualizacao(df,aux,tipo):
    if tipo == 1:
        for i in range(df.shape[0]):
            df2 = pd.DataFrame();
            if len(aux[i]) != 0:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=aux[i].index[-2]);
                aux[i] = pd.concat([aux[i],df2]);
            else:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=date.today()-BDay(1));
                aux[i] = pd.concat([aux[i],df2]);
            aux[i].dropna(subset=['Open'],inplace = True)
            aux[i] = aux[i][~aux[i].index.duplicated(keep='last')]
            df.iat[i,tipo] = aux[i];
    if tipo == 2:
        for i in range(df.shape[0]):
            df2 = pd.DataFrame();
            if len(aux[i]) != 0:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=aux[i].index[-2],interval="1wk");
                aux[i] = pd.concat([aux[i],df2]);
            else:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=date.today() - datetime.timedelta(days=7),interval="1wk");
                aux[i] = pd.concat([aux[i],df2]);
            aux[i].dropna(subset=['Open'],inplace = True)
            aux[i] = aux[i][~aux[i].index.duplicated(keep='last')]
            df.iat[i,tipo] = aux[i];
    if tipo == 3:
        for i in range(df.shape[0]):
            df2 = pd.DataFrame();
            if len(aux[i]) != 0:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=aux[i].index[-2],interval="1mo");
                aux[i] = pd.concat([aux[i],df2]);
            else:
                df2 = pdr.get_data_yahoo(df.iloc[i,0]+'.sa',start=date.today() - datetime.timedelta(days=30),interval="1mo");
                aux[i] = pd.concat([aux[i],df2]);
            aux[i].dropna(subset=['Open'],inplace = True)
            aux[i] = aux[i][~aux[i].index.duplicated(keep='last')]
            df.iat[i,tipo] = aux[i];
    return(df,aux)


#atualização
def atualizei(tipo):
    df = pd.read_hdf('df.h5', 'obj1');
    aux = []
    for i in range(df.shape[0]-1,-1,-1):
        if df.iloc[i,1].shape[0] <= 3:
            df = df.drop(df.index[i])
    df.index = range(df.shape[0])
    if tipo == 1:
        for i in range(df.shape[0]):
            aux.append(df.iloc[i,tipo])
        df,aux = atualizacao(df,aux,tipo)
        df,aux = MME9(df,aux,tipo)
        df,aux = tendencia(df,aux,tipo)
        
    if tipo == 2:
        for i in range(df.shape[0]):
            aux.append(df.iloc[i,tipo])
        df,aux = atualizacao(df,aux,tipo)
        df,aux = MME9(df,aux,tipo)
        df,aux = tendencia(df,aux,tipo)
        
    if tipo == 3:
        for i in range(df.shape[0]):
            aux.append(df.iloc[i,tipo])
        df,aux = atualizacao(df,aux,tipo)
        df,aux = MME9(df,aux,tipo)
        df,aux = tendencia(df,aux,tipo)
        
    salvardf(df);
    
    return()



def LAD():
    LADdf = pd.read_hdf('LAD.h5', 'obj1')
    a = datetime.datetime.now();
    lad = 0;
    if a.hour >= 17:
        for i in range(df.shape[0]):
            if aux[i].Close[aux[i].shape[0]-1] > aux[i].Open[aux[i].shape[0]-1]:
                lad += 1
            if aux[i].Close[aux[i].shape[0]-1] < aux[i].Open[aux[i].shape[0]-1]:
                lad -=1
    
    LADdf.append(lad)
    LADdf.to_hdf('LAD.h5', 'obj1');
    return()

def delcol(df,aux):
    for i in range(df.shape[0]):
        aux[i].drop(columns=['MME9'],errors='ignore',inplace = True)
    return(aux)


def salvardf(df):
    df.to_hdf('df.h5', 'obj1');
    return

def carregardf():
    df = pd.DataFrame();
    aux = [];
    auxwk = [];
    auxmo = [];
    df = pd.read_hdf('df.h5', 'obj1');
    for i in range(df.shape[0]-1,-1,-1):
        if df.iloc[i,1].shape[0] <= 3:
            df = df.drop(df.index[i])
    df.index = range(df.shape[0])
    for i in range(df.shape[0]):
        aux.append(df.iloc[i,1])
        auxwk.append(df.iloc[i,2])
        auxmo.append(df.iloc[i,3])
    return(df,aux,auxwk,auxmo)

def index(ticker):
    index = (df[df['tickers'] == ticker].index.values)
    return index
```
